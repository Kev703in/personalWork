# Filling date to all time-slots

```python
dated_df = df.copy()
for index, rows in dated_df.iterrows():
    if rows['date'] != '-' :       
        date = rows['date']
    else:
        dated_df.at[index,'date'] = date
    
dated_df.head(10)
```

# Ratio of ad clicks over avg time spent watching ads

```python
sum = 0
for index, rows in dated_df.iterrows():
	sum = sum + (rows['ads_clicked']/rows['total_ads_watched_in_mins'])

sum/len(dated_df)
```

# Max ad screen time ratio per date

```python
new = dated_df.groupby('date')[['total_user_time_spent_in_mins','total_ads_watched_in_mins']].sum().reset_index()
new['ads_time_on_plt_date'] = new['total_ads_watched_in_mins']/new['total_user_time_spent_in_mins']
new.iloc[19:25]
```

```python
print("Date with Max ad screen time ratio :",new.loc[new['ads_time_on_plt_date'].idxmax()]['date'])
```

# Max ad screen time ratio per timeslot

```python
new = dated_df.groupby('time_slot')[['total_user_time_spent_in_mins','total_ads_watched_in_mins']].sum().reset_index()
new['ads_time_on_plt_time'] = new['total_ads_watched_in_mins']/new['total_user_time_spent_in_mins']
new.head(5)
```

```python
print("Time_slot with Max ad screen time ratio :", new.loc[new['ads_time_on_plt_time'].idxmax()]['time_slot'])
```

# min viewership in week A:

```python
dated_df['date'] = pd.to_datetime(dated_df['date'])

# Adding Day of week column with Day name
dated_df["day_of_week"] = dated_df['date'].dt.day_name()

# Segregating dates into Weekly periods
dated_df["week"] = dated_df['date'].dt.to_period('W')
```

```python
new = dated_df.groupby('week')[['total_user_time_spent_in_mins','total_ads_watched_in_mins','ads_clicked','user_time_spent_versionB_in_mins','ads_watched_vesionB_in_mins','ads_clicked_versionB']].sum().reset_index()
new
```

```python
print("min viewership in week A: ",new.loc[new['total_user_time_spent_in_mins'].idxmin()]['week'])
```
