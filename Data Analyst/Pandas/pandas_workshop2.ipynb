# Number of platform_ids from ads_statistics.csv not in platforms.csv

```python
new = df_ads_statistics.copy()
new["in_platforms"] = df_ads_statistics["platform_id"].isin(df_platforms["platform_id"])
(new["in_platforms"] == False).sum()
```

# ad name for which most money spent broadcasting

```python
new = df_ads_statistics.copy()
new['amount'] = new['watch_count']*new['price_per_watch']
new = new.groupby('video_id')[['amount']].sum().reset_index()
new
```

```python
new1 = df_ads.copy()
new1
```

```python
new1['long_cost'] = np.nan
new1['short_cost'] = np.nan
for index,rows in new1.iterrows():
    long_cost = new.loc[new['video_id']==rows['long_version_video_id'],'amount']
    short_cost = new.loc[new['video_id']==rows['short_version_video_id'],'amount']
    if not long_cost.empty:
        new1.at[index,'long_cost']= long_cost.values[0]
    if not short_cost.empty:
        new1.at[index,'short_cost']= short_cost.values[0]
new1
```

```python
# There are some video_ids in ads dataset which are not in ads_statistics dataset
new1[new1.isnull().any(axis=1)]
```

```python
# nan values to 0
new1["long_cost"] = new1["long_cost"].fillna(0)
new1["short_cost"] = new1["short_cost"].fillna(0)
new1["total_cost"] = new1["long_cost"] + new1["short_cost"]
```

```python
new1.loc[new1['total_cost'].idxmax()]
```
