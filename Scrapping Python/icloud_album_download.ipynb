```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import requests
```

```python
# Set up Chrome options to handle file downloads
chrome_options = Options()
download_dir = "C:/Users/kev70/OneDrive/Desktop/try/downloads"  # Change this to your desired download directory

chrome_options.add_experimental_option("prefs", {
    "download.default_directory": download_dir,
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
})

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)
time.sleep(5)
```

```python
#Enter album url
url = "enter album url" 
driver.get(url)
time.sleep(1)

photo_elements = driver.find_elements(By.XPATH, "//div[@class='view x-stream-photo-group-blocks-container-view']")  
photo_elements[0].click()
time.sleep(1)
downloaded_files=[]
flag = True
count = 0

while flag:
    iframe = driver.find_element(By.XPATH, "//img[@src]")
    src_url = iframe.get_attribute('src')

    print(f"Found iframe src URL: {src_url}")
    driver.switch_to.default_content()
    response = requests.get(src_url)
    file_name = src_url.split("/")[-1].split("?")[0]
    file_path = f"{download_dir}/{file_name}"
    print(file_name)
    if file_name in downloaded_files:
        count = count + 1
        if count >= 10:
            flag = False
    else:
        with open(file_path, 'wb') as file:
            file.write(response.content)
            downloaded_files.append(file_name)
        time.sleep(1)
        slideshow_next = driver.find_elements(By.XPATH, "//div[@class='view slideshow-item photo']")  
        slideshow_next[0].click()
        time.sleep(1)
```
